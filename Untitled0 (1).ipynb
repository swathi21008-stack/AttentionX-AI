```python


client = Groq(api_key="gsk_5txxxxxxxxxxxxxxxxxxxxxx")

def analyze_video_transcript(transcript):
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": "You are a video expert. Analyze this transcript and find TOP 3 engaging moments with timestamps and captions: " + transcript
            }
        ],
        model="llama-3.3-70b-versatile",
    )
    return chat_completion.choices[0].message.content

sample_transcript = """
0:00 - Hello everyone, welcome to my channel
0:30 - Today I will show you how to make money online
1:00 - This one trick changed my life completely
2:00 - I went from zero to 10000 dollars in one month
3:00 - Thanks for watching, subscribe for more
"""

result = analyze_video_transcript(sample_transcript)
print(result)
```
